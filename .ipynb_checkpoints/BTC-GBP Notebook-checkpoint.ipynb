## (a) (i) Plot a histogram of the daily returns over the last few years

```python
returns = []
for day in range(len(adj_close)-1):
    returns.append(np.log((adj_close[day+1] / adj_close[day])))
```

```python
m, s = stats.norm.fit(returns) # get mean and standard deviation 
```

```python
m
```

```python
s
```

```python
# Visualize histogram with overlaid probability distributions
plt.figure(figsize=(5,5))
plt.rcParams.update({'font.size': 14})
plt.hist(returns, bins=len(returns), normed=True)#, normed=True)
lnspc1 = np.linspace(-0.4, 0.4, len(returns))
pdf_g1 = stats.norm.pdf(lnspc1, m, s) 
plt.plot(lnspc1, pdf_g1, label="Normal ($\mu = 0.004$, $\sigma = 0.07$)") # plot it
pdf_cauch = stats.cauchy.pdf(lnspc1, m, 0.02)  # This gamma of 0.02 was chosen totally randomly
plt.plot(lnspc1, pdf_cauch, label="Cauchy ($\mu = 0.004$, $\gamma = 0.02$)", linewidth=2.5) # plot it
plt.legend()
plt.ylabel('Density')
plt.xlabel('Daily Return')
plt.xlim(-0.3,0.3)
plt.savefig('plot', dpi = 300, bbox_inches='tight')
plt.show()
```

## (b) (i) Calculate the annual drift and volatility of the share price using a 60 day rolling average over the last year, plotting your results on a graph as a function of time. 

### Drift $\mu$

$ U = \ln{\frac{S_{i}}{S_{i-1}}}$ 

$\mu = \frac{{<U_{i}>}+ \frac{\sigma^{2}}{2}}{\Delta t}$

## Plot the drift using a 60 day rolling average.

i.e. Plot the drift at each day, taking an average of the 60 days before that day.

```python
last_year_adj_close = list(data['Adj Close'][-365-60:])
```

```python
# Get the returns as a list
t = 0  # Corresponds to day -60.
dt = 1
mu_list = []
t_list = list(range(365))
S_i_list = []
previous_S_i = last_year_adj_close[0]

for i in range(len(last_year_adj_close)):
    S_i = last_year_adj_close[i]  # Current share price in time
    S_i_list.append(S_i)
    mu_i = np.log(S_i / (previous_S_i))  # Daily returns
    mu_list.append(mu_i)
    previous_S_i = S_i
```

```python
# Get the volatility and standard deviation
t0 = 0
std_list = []
mean_list = []
volatility_list = []
while t0 < 365:
    m, s = stats.norm.fit(mu_list[t0:t0+60])
    std_list.append(s)
    mean_list.append(m)
    t0 += 1
    volatility = s / np.sqrt(1/365)
    volatility_list.append(volatility)
```

```python
# Get the drift
t0 = 0
drift_list = []
for i in range(365):
    take_average_of = mu_list[t0:t0+60]    
    drift_list.append(np.mean(take_average_of) + (std_list[i]**2)/2) 
    t0 += 1
```

```python
plt.plot(t_list, drift_list)
plt.show()
```

```python
plt.plot(t_list, volatility_list)
plt.show()
```

## Repeat, but using MAD

```python
t0 = 0
drift_list_mad = []
for i in range(365):
    take_average_of = mu_list[t0:t0+60]    
    drift_list_mad.append(np.median(take_average_of) + (std_list[i]**2)/2) 
    t0 += 1
```

```python
t0 = 0
MAD = []
volatility_mad = []
while t0 < 365:
    data = mu_list[t0:t0+60]
    take_median_of = []
    for i in range(len(data)):
        take_median_of.append(abs(data[i] - np.median(data))) 
    t0 += 1
    MAD = 1.4826 * np.median(take_median_of)
    volatility_mad.append(MAD / np.sqrt(1/365))
```

```python
plt.plot(t_list, volatility_mad)
plt.show()
```

```python
plt.plot(t_list, drift_list_mad)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.rcParams.update({'font.size': 14})
plt.plot(drift_list, label='mean')
plt.plot(drift_list_mad, label='median')
# plt.title('Drift of Bitcoin Since the Past Year, using Median')
plt.ylabel('Drift')
plt.xlabel('Days')
plt.legend()
plt.savefig('plot', dpi = 300, bbox_inches='tight')
plt.show()
```

# (c) (i) Using LIBOR as the risk free borrowing rate calculate the value of 1 month and 6 month at-the-money call option, stating any assumptions you make. 

## Assumptions

* No transaction fees
* No divedends, i.e. q=0

You should mention that the trading days for bitcoin in a year are 365, as it nevers stops.

```python
from scipy.stats import norm
```

Price of at-the-money call option for __1 month__:

```python
S0 = last_year_adj_close[-1]
k = S0
sigma = np.mean(volatility_mad)
r = 0.04  # LIBOR risk free rate
T = 1/12  # 1 month

z_plus = ((r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))
z_minus = ((r - sigma**2 / 2) * T) / (sigma * np.sqrt(T))

C = S0 * (norm.cdf(z_plus) - np.exp(-r * T) * norm.cdf(z_minus))

C
```

Price of at-the-money call option for __6 months__:

```python
S0 = last_year_adj_close[-1]
k = S0
sigma = np.mean(volatility_mad)
r = 0.04  # LIBOR risk free rate
T = 6/12  # 6 months

z_plus = ((r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))
z_minus = ((r - sigma**2 / 2) * T) / (sigma * np.sqrt(T))

C = S0 * (norm.cdf(z_plus) - np.exp(-r * T) * norm.cdf(z_minus))
C
```
